```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linearReg(x,y):
    SSxy = np.sum(x*y) - np.mean(y)*x
    SSxx = np.sum(x*x) - np.mean(x)*x
    
    B = SSxy/SSxx
    a = np.mean(y) - B*np.mean(x)
    Y = a + B*x
    
    plt.scatter(x, y, label='Actual Values')
    plt.plot(x, Y, label='Predicted Values', color = 'r')
    
    plt.title('Linear Regression Graph')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

def main():
    x = np.arange(1,11)
    y = np.array([1, 3, 2, 5, 7, 8, 8, 9, 10, 12])
    linearReg(x,y)
    
if __name__ == "__main__":
    main()    
```
